```python
import os
import matplotlib.pyplot as plt
import numpy as np
import re  # For regex to extract and process file names

# Folder containing the CSV files
folder_path = './lepton-lepton'  # Update with your folder path

# Define line styles and a rich color palette
line_styles = ['-', '--', '-.', ':']
color_map = plt.colormaps['tab20']  # Updated colormap access method
group_styles = {}  # To store styles for each V_n group
current_style_index = 0
current_color_index = 0

# Initialize the main plot
plt.figure(figsize=(14, 6))

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):  # Process only CSV files
        file_path = os.path.join(folder_path, filename)
        
        # Extract data from the file
        x_data = []
        y_data = []
        with open(file_path, 'r') as f:
            for line in f:
                x, y = map(float, line.strip().split(','))  # Assuming two columns separated by a comma
                x_data.append(x)
                y_data.append(y)
        
        # Process legend label
        match = re.match(r'(\d+)([A-Za-z].+)', filename.replace('.csv', ''))
        if match:
            group_number = match.group(1)  # Leading number (e.g., '2' for '2Almasi2020')
            rest_of_label = match.group(2)  # Remaining part of the filename (e.g., 'Almasi2020_m_abs_ee')
            group = f"V{group_number}"  # Group name (e.g., 'V2')
            if group not in group_styles:  # Assign a unique style for the group
                group_styles[group] = line_styles[current_style_index]
                current_style_index = (current_style_index + 1) % len(line_styles)

            # Apply rules for legend name
            if "astro" in filename:  # Rule for astro case
                parts = filename.split('astro')
                if len(parts) > 1:
                    prefix = parts[0].split('_')[-1]  # Extract the part before "astro"
                    label = f"astro-{prefix}"
                else:
                    label = "astro"
            else:  # General case
                parts = rest_of_label.split('_')
                if len(parts) > 3:
                    label = f"{group}_{parts[0]}_{parts[-1]}"  # Keep only specific parts
                else:
                    label = f"{group}_{rest_of_label}"  # Fallback if format is unexpected

            line_style = group_styles[group]  # Get the line style for this group
        else:
            # Default behavior if no leading number is found
            label = filename.replace('.csv', '')  # Use full filename without '.csv'
            line_style = '-'  # Default to solid line for non-grouped files

        # Assign a unique color for this line
        color = color_map(current_color_index)
        current_color_index = (current_color_index + 1) % color_map.N  # Cycle through colors if needed

        # Plot the data with the group's line style and unique color
        plt.plot(x_data, y_data, label=label, linestyle=line_style, color=color, linewidth=2)

# Customize the main plot
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e-15, 1e11)  # Set x-axis range
plt.ylim(1e-50, 1e-2)  # Set y-axis range
plt.xlabel(r'$\lambda$ (m)', fontsize=14)
plt.ylabel(r'$g_Ag_A$', fontsize=14)
plt.title('Limits for Axial-vector/Axial-vector interactions', fontsize=16)

# Add the main legend (line colors and labels)
plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), fontsize=10, title='Data')

# Add grid
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Adjust layout to make space for the legend
plt.tight_layout(rect=[0, 0, 0.8, 1])

# Show the main plot
plt.show()

# Define group styles first
group_styles = {
    "$V_3$": "-",
    "$V_8$": "--",
    "$V_2$": "-.",
    "Astro": ":"
}

# Create a second figure for the line style legend
plt.figure(figsize=(6, 2));
#plt.title('Line Styles for Groups', fontsize=14)

# Add a dummy line for each group using the correct styles
for group, style in group_styles.items():
    plt.plot([], [], linestyle=style, color='black', linewidth=2, label=group);

# Add the legend for the line styles
plt.legend(loc='center', fontsize=10, ncol=len(group_styles));

# Hide axes and ensure the focus is on the legend
plt.axis('off')
plt.tight_layout()
plt.show()  # Properly renders the second figure
```
